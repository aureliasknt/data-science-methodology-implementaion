#2. Identifikasi Masalah

Indeks Prestasi Kumulatif (IPK) menjadi acuan berhasilnya mahasiswa dalam perkuliahan pada perguruan tinggi dalam negeri maupun luar negeri. Apabila mahasiswa memiliki IPK di atas rata-rata, perguruan tinggi dikatakan telah berhasil dalam mendidik serta mengelola institusi. Tidak hanya itu, indikator lainnya yang membuat perguruan tinggi dikatakan berhasil adalah jika alumninya mampu memenuhi keinginan stakeholder (pihak pihak yang berkepentingan/para pengguna) dengan salah satu indikator alumni diserap industri atau dunia kerja.
Saat ini  dalam masyarakat dan lembaga Perguruan Tinggi (PT) cenderung hanya terdapat mahasiswa dengan kemampuan akademik yang tinggi, tanpa diimbangi dengan pembentukan karakter yang kuat dan cerdas [1]. Berdasarkan pernyataan tersebut, seharusnya kemampuan hard skills seorang mahasiswa (yang diwakili dengan IPK) harus juga diimbangi dengan kemampuan soft skills. Hal ini dapat disampaikan karena rasio kebutuhan soft skills dan hard skills di dunia kerja berbanding terbalik dengan pengembangan soft skills di perguruan tinggi, yang mana perguruan tinggi lebih mementingkan pengembangan  hard skills dibandingkan soft skills. Padahal, hal yang dibutuhkan untuk membawa dan mempertahankan seseorang dalam sebuah kesuksesan didunia kerja adalah  80% soft skills dan 20% hard skills [2].   

Untuk mengembangkan soft skills, mahasiswa dapat mengikuti kegiatan pembelajaran (intrakurikuler) dan kegiatan kemahasiswaan (ekstrakurikuler). Pengembangan soft skill melalui kegiatan pembelajaran atau di kelas tatap muka membutuhkan kreativitas guru mata pelajaran dengan tetap  menjaga pencapaian kompetensi  mata pelajaran. Mengembangkan soft skills melalui kurikulum dapat ditempuh dengan dua cara. Pertama, melalui kegiatan pembelajaran yang secara khusus diintegrasikan ke dalam mata kuliah sesuai Kurikulum, SAP (Unit Program Kuliah) atau RPS (Rencana Pembelajaran Semester). Kedua, dapat dilakukan melalui proses hidden curriculum, yaitu  strategi untuk mengembangkan soft skills yang diberikan dosen kepada mahasiswa  secara terpadu selama perkuliahan. Mahasiswa juga dapat mengembangkan soft skill melalui kegiatan kemahasiswaan (ekstrakurikuler) berupa mengikuti organisasi kemahasiswaan, perlombaan, kepanitiaan,dan lain lain.  Hal ini didukung dengan hasil penelitian mengenai penerapan model pembelajaran kooperatif dapat meningkatkan soft skills dan hard skills mahasiswa dan mahasiswa memberikan respon positif terhadap model pembelajaran yang diterapkan [3].

### Penanganan *Missing Value*

Penanganan missing value ini dilakukan untuk membersihkan data dari data yang hilang (missing value) agar data lebih mudah dianalisis. Dalam menangani missing value ini, data di imputasi atau mengganti nilainya

```python
data_copy.head(10)
```

```python
import numpy as np
data_copy = data_copy.replace (np.NaN, 0.0)

def imputasi (df_input) :
  list_columns = df_input.columns
  class_column = list_columns[-1] 

  for column in list_columns [:-1] : 
     if(column=="total_posts" or column=="helpful_post" or column=="nice_code_post" or column=="collaborative_post" or column=="confused_post"
     or column=="creative_post" or column=="bad_posts" or column=="amazing_posts" or column=="timeonline"):
        df_input[column] = df_input[column].fillna(df_input.groupby(class_column) [column].transform('mean'))
  return df_input

data_imputasi = imputasi(data_copy)
data_imputasi.isnull().values.any()
```

### Normalisasi *MinMax*

Dilakukannya normalisasi bertujuan untuk  menghilangkan redundansi data (pengulangan) dan menstandarisasi informasi untuk alur kerja data yang lebih baik. Pada kasus ini kami menggunakan metode *MinMax* untuk melakukan normalisasi data.

Persamaan dari fungsi normalisasi *MinMax* adalah sebagai berikut:

$x'=\frac{x-min(x)}{max(x)-min(x)}$

Dengan rumus tersebut, lakukan normalisasi *dataset* Anda dengan metode normalisasi *MinMax*.



####Rentang data sebelum dilakukan normalisasi

```python
def cetak_rentang (df_input) : 
  list_fitur = df_input.columns [:-1]
  
  for fitur in list_fitur : 
    max = df_input[fitur].max()
    min = df_input[fitur].min()
    print ("Rentang Fitur: ",fitur, "adalah ",max-min)
```

```python
cetak_rentang(data_sk1)
```

```python
list_fitur_sk1 = data_sk1.columns[:-1]
list_fitur_sk2 = data_sk2.columns[:-1]
list_fitur_sk3 = data_sk3.columns[:-1]
list_fitur_sk4 = data_sk4.columns[:-1]
list_fitur_sk5 = data_sk5.columns[:-1]
```

```python
list_fitur_sk1
```

####Hasil normalisasi

```python
def minmax(data_input):
  list_fitur = data_input.columns[:-1]
  for fitur in list_fitur:
    max = data_input[fitur].max()
    min = data_input[fitur].min()
    data_input[fitur] = (data_input[fitur]-min)/(max-min)
    return data_input

data_normalized_sk1 = minmax(data_sk1)
data_normalized_sk2 = minmax(data_sk2)
data_normalized_sk3 = minmax(data_sk3)
data_normalized_sk4 = minmax(data_sk4)
data_normalized_sk5 = minmax(data_sk5)
```
